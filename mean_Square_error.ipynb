# What is Mean Squared Error (MSE)?
Mean Squared Error is a measure of the average squared differences between the actual values and the predicted values. It gives us an idea of how close our model's predictions are to the actual values.

# How is MSE Calculated?
To calculate MSE, follow these steps:

1. Compute the Difference: Subtract the predicted value from the actual value for each data point.
2. Square the Difference: Square each difference to ensure all values are positive.
3. Average the Squares: Take the mean of these squared differences.

# Way to find  MSE
1. Using numpy
2. sklearn

```python
y_actual = np.array([2,3,5,1,5])
y_pred = np.array([2.2,3.1,4.6,1.2,4.5])


MSE = (np.sum(np.square(y_pred - y_actual)))
MSE = MSE/y_actual.size

print("Mean squared error:\n", MSE)
```

```python
mse = np.mean((y_pred - y_actual) ** 2)
print("Mean Squared Error:", mse)
```

# 2. Using Sklearn

```python
from sklearn.metrics import mean_squared_error
```

```python
y_actual = np.array([2,3,5,1,5])
y_pred = np.array([2.2,3.1,4.6,1.2,4.5])

MSE = mean_squared_error(y_actual, y_pred)
print("mean_squared_error: ", MSE)
```

# Why is MSE Important?
Penalty for Large Errors: Squaring the differences means larger errors are penalized more, making MSE sensitive to outliers.
1. Model Comparison: MSE allows us to compare different models and choose the one with the lowest error.
2. Optimization: Many algorithms, like linear regression, minimize MSE to find the best fit line.

# Practical Considerations
1. Scale: MSE is in the squared units of the response variable, which can be difficult to interpret directly. Root Mean Squared Error (RMSE) is often used for more interpretable units.
2. Outliers: Due to the squaring operation, MSE can be heavily influenced by outliers. Always consider the nature of your data when interpreting MSE.
