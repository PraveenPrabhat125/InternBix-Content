# Function
Understanding and leveraging different types of functions in Python is key to writing efficient and readable code. Here are some essential function types you should know:

## Key types of function in pythons
1. User define function
2. Anonymous Functions (Lambda Functions)
3. Built-in Functions
4. Higher-Order Functions
5. Recursive Functions

# 1. User define function

```python
def user_def_fun(name):
  print(f"Hello, lets learn functin {name}")

# calling function
user_def_fun('InternBix')
```

# 2. Anonymous Functions (Lambda Functions)

```python
x = 10
y = 6

add = lambda x, y : x * y

print(add(x,y))
```

# 3. Built-in Functions
Description: Functions that are always available in Python.
How: Examples include print(), len(), range().

```python
print('Hello, World!')
```

# 4. Higher-Order Functions

```python
def apply_funct(funct, val):
  return funct(val)

def square(x):
  return x*x

apply_funct(square, 5)
```

# 5. Recursive Functions
Description: Functions that yield items one at a time, creating an iterator.
How: Defined using the yield keyword.

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

for number in count_up_to(5):
    print(number)  # Output: 1 2 3 4 5
```

### Other way
The count_up_to function itself does not return a final value in the traditional sense (like a return statement would). Instead, it returns a generator object that can be used to generate a sequence of values on-the-fly, which can be highly memory-efficient for working with large datasets or streams of data.

```python
def count_up_to(max):
    count = 1
    while count <= max:
        yield count
        count += 1

gen = count_up_to(5)
print(gen)
```

```python
for number in gen:
    print(number)
```
