# Grouping and data aggrigation

```python
# Sample data
data = {
    'Product': ['A', 'B', 'A', 'B', 'A', 'C'],
    'Category': ['Electronics', 'Electronics', 'Electronics', 'Furniture', 'Furniture', 'Furniture'],
    'Sales': [100, 200, 150, 250, 300, 100],
    'Quantity': [1, 2, 3, 1, 2, 3]
}

df = pd.DataFrame(data)
print(df)
```

```python
p_gp = df.groupby('Product')['Sales'].sum()
print(p_gp)
```

```python
prod_cat = df.groupby(['Product', 'Category'])['Sales'].sum()
prod_cat
```

- Applying Multiple Aggregations
You can also apply multiple aggregation functions to the grouped data. For example, to calculate both the sum and average sales for each category:

```python
# Example data
data = {
    'Product': ['A', 'A', 'B', 'B', 'C', 'C'],
    'Category': ['X', 'Y', 'X', 'Y', 'X', 'Y'],
    'Sales': [100, 150, 200, 250, 300, 350]
}

df = pd.DataFrame(data)

# Group by 'Product' and 'Category', and aggregate 'Sales'
result = df.groupby(['Product', 'Category']).agg(
    total_sales=('Sales', 'sum'),
    average_sales=('Sales', 'mean'),
    max_sales=('Sales', 'max')
)

print(result)
```

```python
# df.head(3)
summary1 = df.groupby(['Product', 'Category']).agg(
    total_sales = ('Sales', 'sum'),
    Max_sales = ('Sales', 'max'),
    Avg_sales = ('Sales', 'mean')
)
summary1
```

```python
df['log_sales'] = df['Sales']/10
df.head()
```

# 9. MErging and Joining DFs

```python
df1 = pd.DataFrame({'key': ['A', 'B', 'C'], 'value': [1, 2, 3]})
df2 = pd.DataFrame({'key': ['B', 'C', 'D'], 'value': [4, 5, 6]})

# inner merge
merged_df = pd.merge(df1, df2, on = 'key', how = 'inner')
print('Inner join\n', merged_df)


# left merge
merged__left_df = pd.merge(df1, df2, on = 'key', how = 'left')
print('Left join\n', merged__left_df)
```

# Joining DF

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 'B': ['B0', 'B1', 'B2']})
df2 = pd.DataFrame({'C': ['C0', 'C1', 'C2'], 'D': ['D0', 'D1', 'D2']}, index=[1, 2, 3])

print('Df1: \n', df1)
print('Df2: \n', df2)

joined_df = df1.join(df2)
print(joined_df)
```
