#Joins in Python
Joins in Python are primarily performed using the Pandas library, which offers powerful and flexible data structures like DataFrame to handle data manipulation and analysis. Let's explore the different types of joins: inner, left, right, outer, and cross joins.

# 1. Inner Join
An inner join returns only the rows with matching values in both DataFrames.

image.png

```python
import pandas as pd

df1 = pd.DataFrame({'key': ['A', 'B', 'C', 'D'],
                    'value_df1': [1, 2, 3, 4]})


df2 = pd.DataFrame({'key': ['B', 'C', 'D', 'E'],
                    'value_df2': [5, 6, 7, 8]})
print('df1: \n', df1)
print('df2: \n', df2)
```

```python
Inner_joins = pd.merge(df1, df2, on = 'key', how = 'inner')
Inner_joins
```

# 2.Left Join
A left join returns all rows from the left DataFrame, and the matched rows from the right DataFrame. Unmatched rows will have NaN for columns from the right DataFrame.

image.png

```python
left_joined = pd.merge(df1, df2, on = 'key', how = 'left')
left_joined
```

# 3. Right Join
A right join returns all rows from the right DataFrame, and the matched rows from the left DataFrame. Unmatched rows will have NaN for columns from the left DataFrame.

image.png

```python
right_joined = pd.merge(df1, df2, on = 'key', how = 'right')
right_joined
```

# 4. Outer Join
An outer join returns all rows when there is a match in either left or right DataFrame. Unmatched rows will have NaN for columns from the other DataFrame.

image.png

```python
outer_join = pd.merge(df1, df2, on='key', how='outer')
print(outer_join)
```

# 5. Cross Join
A cross join returns the Cartesian product of the two DataFrames, i.e., all possible combinations of rows.

image.png

```python
cross_join = df1.merge(df2, how = 'cross')
cross_join
```

# Summary of Different Joins
1. Inner Join: Returns rows with matching keys in both DataFrames.
2. Left Join: Returns all rows from the left DataFrame and matching rows from the right.
3. Right Join: Returns all rows from the right DataFrame and matching rows from the left.
4. Outer Join: Returns all rows from both DataFrames, filling NaNs for non-matching keys.
5. Cross Join: Returns the Cartesian product of both DataFrames.
